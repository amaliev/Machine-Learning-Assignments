# Generating the Data

```python
import numpy as np;
import matplotlib.pyplot as plt;
mu1 = [1,0];
mu2 = [-1,0];
mu3 = [0,-10];
U = [[1/np.sqrt(2),-1/np.sqrt(2)],[1/np.sqrt(2),1/np.sqrt(2)]];
lamb1 = [[4,0],[0,1]];
lamb2 = [[1,0],[0,5]];
sigma1 = np.matmul(np.matmul(U,lamb1),np.transpose(U));
sigma2 = np.matmul(np.matmul(U,lamb2),np.transpose(U));
sigma3 = np.identity(2);
X1 = np.random.multivariate_normal(mu1,sigma1,1000);
X2 = np.random.multivariate_normal(mu2,sigma2,1000);
X3 = np.random.multivariate_normal(mu3,sigma3,1000);
plt.scatter(X1[:,0],X1[:,1],c='r');
plt.scatter(X2[:,0],X2[:,1],c='g');
plt.scatter(X3[:,0],X3[:,1],c='b');
```

# k-Means Clustering

```python
X = np.vstack((X1,X2,X3));
np.random.shuffle(X);
C1 = X[0:1000,:];
C2 = X[1000:2000,:];
C3 = X[2000:3000,:];
done = False;
while (done == False) :
    done = True;
    size1 = np.shape(C1)[0];
    size2 = np.shape(C2)[0];
    size3 = np.shape(C3)[0];
    mu1 = np.sum(C1,0) / size1;
    mu2 = np.sum(C2,0) / size2;
    mu3 = np.sum(C3,0) / size3;
    i = 0;
    while i < np.shape(C1)[0] :
        d1 = np.linalg.norm(C1[i,:]-mu1);
        d2 = np.linalg.norm(C1[i,:]-mu2);
        d3 = np.linalg.norm(C1[i,:]-mu3);
        newC = np.argmin([d1,d2,d3]);
        if (newC == 1) : 
            done = False;
            C2 = np.vstack((C2,C1[i,:]));
            C1 = np.delete(C1,i,0);
        if (newC == 2) :
            done = False;
            C3 = np.vstack((C3,C1[i,:]));
            C1 = np.delete(C1,i,0);
        if (newC == 0) :
            i = i+1;
    j = 0;
    while j < np.shape(C2)[0] :
        d1 = np.linalg.norm(C2[j,:]-mu1);
        d2 = np.linalg.norm(C2[j,:]-mu2);
        d3 = np.linalg.norm(C2[j,:]-mu3);
        newC = np.argmin([d1,d2,d3]);
        if (newC == 0) : 
            done = False;
            C1 = np.vstack((C1,C2[j,:]));
            C2 = np.delete(C2,j,0);
        if (newC == 2) : 
            done = False;
            C3 = np.vstack((C3,C2[j,:]));
            C2 = np.delete(C2,j,0);
        if (newC == 1) : 
            j = j+1;
        j = j+1;
    k = 0;
    while k < np.shape(C3)[0] :
        d1 = np.linalg.norm(C3[k,:]-mu1);
        d2 = np.linalg.norm(C3[k,:]-mu2);
        d3 = np.linalg.norm(C3[k,:]-mu3);
        newC = np.argmin([d1,d2,d3]);
        if (newC == 0) : 
            done = False;
            C1 = np.vstack((C1,C3[k,:]));
            C3 = np.delete(C3,k,0);
        if (newC == 1) : 
            done = False;
            C2 = np.vstack((C2,C3[k,:]));
            C3 = np.delete(C3,k,0);
        if (newC == 2) :
            k = k+1;
print("The sizes of the clusters are: ",np.shape(C1)[0]," ",np.shape(C2)[0]," ",np.shape(C3)[0]);
```

```python
plt.scatter(C1[:,0],C1[:,1],c='r');
plt.scatter(C2[:,0],C2[:,1],c='g');
plt.scatter(C3[:,0],C3[:,1],c='b');
```

```python
import matplotlib.pyplot as plt;
import numpy as np;
x = np.linspace(1930,2010,9);
y = [1.015, 1.33, 2.29, 3.26, 4.951, 6.759, 9.366, 13.131, 15.599];
plt.plot(x,y,'ro--');
plt.title('Real US GDP (in trillions)');
plt.xlabel('Year');
plt.ylabel('Trillions of $');
plt.show();
```

There may be a quadratic relationship between GDP and year.  
This is because the rate of change $\frac{dGDP}{dt}$ seems to increase uniformly.  
Note that:  
$$GDP(1930)\sim 1=0^2+1$$
$$GDP(1950)\sim 2=1^2+1$$
$$GDP(1970)\sim 5=2^2+1$$
$$GDP(1990)\sim 10=3^2+1$$
$$GDP(2010)\sim 17=4^2+1$$
Thus, we can roughly model the relationship as $GDP(t)=\left(\frac{t-1930}{20}\right)^2+1$.  

```python
y2 = np.power(np.divide((x-1930),20),2)+1;
plt.plot(x,y,'ro--',x,y2,'b');
plt.title('Real US GDP (in trillions)');
plt.xlabel('Year');
plt.ylabel('Trillions of $');
plt.show();
```

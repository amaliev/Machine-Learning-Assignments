# Gradient Descent

```python
import numpy as np;
def f(w1,w2) :
    return (w1**2+w2-11)**2 + (w1+w2**2-7)**2;
```

```python
def gradient(x,step) :
    grad = [0]*2;
    xn = x[0] + step;
    grad[0] = (f(xn,x[1])-f(x[0],x[1]))/(xn-x[0]);
    xn = x[1] + step;
    grad[1] = (f(x[0],xn)-f(x[0],x[1]))/(xn-x[1]);
    return grad;
```

```python
def newpoint(x,step) :
    grad = gradient(x,step);
    return [x[0]-step*grad[0],x[1]-step*grad[1]];
```

```python
def graddesc(w0,step) :
    i = 0;
    x = w0;
    while (np.log10(np.linalg.norm(gradient(x,step)))>-12) :
        i += 1;
        x = newpoint(x,step);
        if (i > 10000) :
            break;
    if (i>10000) :
        print("Initial point ",w0,"and step size ",step,": timed out.");
    else :
        print("Initial point ",w0,"and step size ",step,": terminated at ",x,"in ",i,"steps.");
```

```python
graddesc([0,-4],0.005);
```

```python
graddesc([0,-4],0.01);
```

```python
graddesc([0,-4],0.01);
graddesc([0.5,-4],0.01);
graddesc([0,4],0.01);
graddesc([0.5,4],0.01);
```

These solutions are all different. Based on the information available, these are all local minima, because we followed the path of gradient descent which reduces the value of the function. If we started at a local maxima, it would terminate immediately, and if we were near a local maxima, it would move away from it.

```python
import matplotlib.pyplot as plt;
y = np.linspace(-5,5,1000);
z = np.zeros((1000,1000));
for i in range(0,999) :
    for j in range(0,999) :
        z[i][j] = f(y[j],y[i]);
fig = plt.figure();
ax = fig.add_subplot(111);
ax.contour(y,y,z,100);
ax.autoscale(False);
w0s = [[0,-4],[0.5,-4],[0,4],[0.5,4]];
for j in range(0,4) :
    w0 = [w0s[j][0],w0s[j][1]];
    x = w0;
    path = [w0];
    i = 0;
    while (np.log10(np.linalg.norm(gradient(x,0.01)))>-12) :
        i += 1;
        x = newpoint(x,0.01);
        path.append(x);
        if (i > 10000) :
            break;
    path = np.array(path);
    ax.plot(path[:,0],path[:,1],'r');
plt.show();
```

# Parameter Estimation

```python
from scipy.stats import multivariate_normal as nrm;
mean = [-1,0,4,1,0.5];
cov = 2*np.identity(5);
N = np.zeros((6,1));
eN = np.zeros((6,1));
for i in range(1,7) :
    i10 = 10**i;
    N[i-1] = i10;
    y = nrm.rvs(mean,cov,i10);
    tmean = np.sum(y,0) / i10;
    err = np.linalg.norm(tmean-mean);
    print("For 10^",i,"samples:");
    print("The average mean vector is ",tmean);
    print("And the error from the actual mean is ",err)
    eN[i-1] = err;
```

```python
plt.loglog(N,eN);
print("The log-log plot of eN versus N is about linearly decreasing.");
print("This implies that N and eN have a negative exponential relationship.");
```
